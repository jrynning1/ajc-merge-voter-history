```python
import pandas as pd
from pathlib import Path
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_colwidth = 999
pd.options.display.width = 999
pd.options.display.max_columns = 999
```

```python
demographics_path = Path().cwd().parent.joinpath('main_voter_dataset/voters_20240410_clean_with_voter_history.csv')
demographics_original = pd.read_csv(demographics_path, dtype=str)
demographics_df = demographics_original.copy()
```

```python
def search_first_last(first_name, last_name):
    return demographics_df[(demographics_df['first_name'] == str(first_name).upper()) & (demographics_df['last_name'] == str(last_name).upper())]
```

```python
search_first_last("Steven", "Ewing")
```
